## Import Data into pandas dataframes

```python

import pandas as pd

```

```python
testdemographic = 'C:/Users/Njoroge Wa Chege/Desktop/Zindi Challenges/Predicting Loan Defaults/data\\testdata\\testdemographics.csv'
testperf = 'C:/Users/Njoroge Wa Chege/Desktop/Zindi Challenges/Predicting Loan Defaults/data\\testdata\\testperf.csv'
testprevloans = 'C:/Users/Njoroge Wa Chege/Desktop/Zindi Challenges/Predicting Loan Defaults/data\\testdata\\testprevloans.csv'

dftest1 = pd.read_csv( testdemographic, header=0)

dftest2 = pd.read_csv( testperf, header=0)

dftest3 = pd.read_csv( testprevloans, header=0)
```

```python
traindemographic = 'C:/Users/Njoroge Wa Chege/Desktop/Zindi Challenges/Predicting Loan Defaults/data\\traindata\\traindemographics.csv'
trainperf = 'C:/Users/Njoroge Wa Chege/Desktop/Zindi Challenges/Predicting Loan Defaults/data\\traindata\\trainperf.csv'
trainprevloans = 'C:/Users/Njoroge Wa Chege/Desktop/Zindi Challenges/Predicting Loan Defaults/data\\traindata\\trainprevloans.csv'

dftrain1 = pd.read_csv( traindemographic, header=0)

dftrain2 = pd.read_csv( trainperf, header=0)

dftrain3 = pd.read_csv( trainprevloans, header=0)
```

```python
df_test_col = pd.merge(dftest1, dftest2,  how='inner', on = 'customerid')#.drop()--->drop the second id

df_test_col = pd.merge(df_test_col, dftest3,how='inner', on = 'customerid')

df_test = df_test_col

df_test.dtypes

```

```python
df_train_col = pd.merge(dftrain1, dftrain2, how='inner', left_on='customerid', right_on = 'customerid')#.drop()--->drop the second id

df_train_col = pd.merge(df_train_col, dftrain3, how='inner', left_on='customerid', right_on = 'customerid')

df_train = df_train_col

```
